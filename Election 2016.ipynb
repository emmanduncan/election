```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, output_file, curdoc
from bokeh.layouts import row, column, widgetbox
from bokeh.models import Button, CustomJS, ColumnDataSource, CheckboxButtonGroup, Panel, Tabs, CheckboxGroup, Range1d

#import seaborn as sns
#import ggplot
#pd.set_option('display.max_rows', 10) #displays up to 10 rows in dataframes
pd.set_option('display.max_columns', 500) #displays up to 500 columns in dataframes
output_notebook() #makes Bokeh graphs display in notebook instead of as a file
```

```python
df = pd.read_csv('http://projects.fivethirtyeight.com/general-model/president_general_polls_2016.csv') #import data as Pandas dataframe
df.head() #displays the dataframe
```

```python
print(df.loc[:,'type'].unique()) #displays unique values of the "type" variable
```

```python
df_po = df[df.loc[:,'type']=='polls-only']
df_po = df_po.reset_index(drop=True) #resets the indexes from original values to 0; drop parameter forgets them
df_po.head()
```

```python
print("Number of rows (polls): " + str(df_po.shape[0]))
print("Number of columns (data categories): " + str(df_po.shape[1]))

print("\nNumber of empty values for each column:")
print(df_po.isnull().sum())
```

```python
df_po.loc[:,'forecastdate'] = pd.to_datetime(df_po.loc[:,'forecastdate'])
df_po.loc[:,'startdate'] = pd.to_datetime(df_po.loc[:,'startdate'])
df_po.loc[:,'enddate'] = pd.to_datetime(df_po.loc[:,'enddate'])

df_po.loc[:, 'forecastdate'].head()
```

```python
df_po.loc[:, ['enddate', 'adjpoll_clinton']].head(10)
```

```python
#Tabs
source = ColumnDataSource({'date':df_po.loc[:, 'enddate'], 'clinton':df_po.loc[:,'adjpoll_clinton'], 
                           'trump':df_po.loc[:, 'adjpoll_trump',], 'johnson':df_po.loc[:, 'adjpoll_johnson'], 
                           'mcmullin':df_po.loc[:, 'adjpoll_mcmullin']})
figProps = dict(plot_width=800, plot_height=500, x_axis_type='datetime', active_scroll='wheel_zoom', 
                x_range=Range1d(min(df_po.loc[:, 'enddate']) - pd.DateOffset(months=1), 
                                max(df_po.loc[:, 'enddate']) + pd.DateOffset(months=1)), 
                y_range=[0,  100], 
                x_axis_label="Date", y_axis_label="Vote Percentage")
dotProps = dict(size=10, alpha=0.04)

#Clinton
p1 = figure(title="Clinton", **figProps)
p1.circle(x='date', y='clinton', source=source, color='blue', **dotProps)
tab1 = Panel(child=p1, title="Clinton")

#Trump
p2 = figure(title="Trump", **figProps)
p2.circle(x='date', y='trump', source=source, color='red', **dotProps)
tab2 = Panel(child=p2, title="Trump")

#Johnson
p3 = figure(title="Johnson", **figProps)
p3.circle(x='date', y='johnson', source=source, color='gold', **dotProps)
tab3 = Panel(child=p3, title="Johnson")

#McMullin
p4 = figure(title="McMullin", **figProps)
p4.circle(x='date', y='mcmullin', source=source, color='green', **dotProps)
tab4 = Panel(child=p4, title="McMullin")

#All
p5 = figure(title="All", **figProps)
p5.circle(x='date', y='clinton', source=source, color='blue', **dotProps)
p5.circle(x='date', y='trump', source=source, color='red', **dotProps)
p5.circle(x='date', y='johnson', source=source, color='gold', **dotProps)
p5.circle(x='date', y='mcmullin', source=source, color='green', **dotProps)
tab5 = Panel(child=p5, title="All")

tabs = Tabs(tabs=[tab5, tab1, tab2, tab3, tab4])
show(tabs)
```
