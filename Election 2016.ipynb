```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.layouts import widgetbox
from bokeh.models.widgets import CheckboxButtonGroup, Panel, Tabs
#import seaborn as sns
#import ggplot
pd.set_option('display.max_columns', 500) #displays up to 500 columns in dataframes
output_notebook() #makes Bokeh graphs display in notebook instead of as a file
```

```python
df = pd.read_csv('http://projects.fivethirtyeight.com/general-model/president_general_polls_2016.csv') #import data as Pandas dataframe
df #displays the dataframe
```

```python
print(df.loc[:,'type'].unique()) #displays unique values of the "type" variable
```

```python
df_po = df[df.loc[:,'type']=='polls-only']
df_po = df_po.reset_index(drop=True) #resets the indexes from original values to 0; drop parameter forgets them
df_po
```

```python
print("Number of rows (polls): " + str(df_po.shape[0]))
print("Number of columns (data categories): " + str(df_po.shape[1]))

print("\nNumber of empty values for each column:")
print(df_po.isnull().sum())
```

```python
df_po.describe()
```

```python
df_po.loc[:,'forecastdate'].head()
```

```python
df_po.loc[:,'forecastdate'] = pd.to_datetime(df_po.loc[:,'forecastdate'])
df_po.loc[:,'startdate'] = pd.to_datetime(df_po.loc[:,'startdate'])
df_po.loc[:,'enddate'] = pd.to_datetime(df_po.loc[:,'enddate'])

df_po.loc[:, 'forecastdate'].head()
```

```python
df_po.loc[:, ['enddate', 'adjpoll_clinton']].head(10)
```

```python
#Tabs
#Clinton
p1 = figure(title="Clinton", plot_width=800, plot_height=500, x_axis_type='datetime', active_scroll='wheel_zoom')
p1.circle(x=df_po.loc[:, 'enddate'], y=df_po.loc[:,'adjpoll_clinton'], size=10, color='blue', alpha=0.03)
tab1 = Panel(child=p1, title="Clinton")

#Trump
p2 = figure(title="Trump", plot_width=800, plot_height=500, x_axis_type='datetime', active_scroll='wheel_zoom')
p2.circle(x=df_po.loc[:, 'enddate'], y=df_po.loc[:,'adjpoll_trump'], size=10, color='red', alpha=0.05)
tab2 = Panel(child=p2, title="Trump")

#Both
p3 = figure(title="Clinton + Trump", plot_width=800, plot_height=500, x_axis_type='datetime', active_scroll='wheel_zoom')
p3.circle(x=df_po.loc[:, 'enddate'], y=df_po.loc[:,'adjpoll_clinton'], size=10, color='blue', alpha=0.03)
p3.circle(x=df_po.loc[:, 'enddate'], y=df_po.loc[:,'adjpoll_trump'], size=10, color='red', alpha=0.03)
tab3 = Panel(child=p3, title="Both")

tabs = Tabs(tabs=[tab1, tab2, tab3])
show(tabs)
```

```python
#CheckboxButtonGroup
#Clinton
p1 = figure(title="Clinton", plot_width=800, plot_height=500, x_axis_type='datetime', active_scroll='wheel_zoom')
p1.circle(x=df_po.loc[:, 'enddate'], y=df_po.loc[:,'adjpoll_clinton'], size=10, color='blue', alpha=0.03)
tab1 = Panel(child=p1, title="Clinton")

#Trump
p2 = figure(title="Trump", plot_width=800, plot_height=500, x_axis_type='datetime', active_scroll='wheel_zoom')
p2.circle(x=df_po.loc[:, 'enddate'], y=df_po.loc[:,'adjpoll_trump'], size=10, color='red', alpha=0.05)
tab2 = Panel(child=p2, title="Trump")

#Both
p3 = figure(title="Clinton + Trump", plot_width=800, plot_height=500, x_axis_type='datetime', active_scroll='wheel_zoom')
p3.circle(x=df_po.loc[:, 'enddate'], y=df_po.loc[:,'adjpoll_clinton'], size=10, color='blue', alpha=0.03)
p3.circle(x=df_po.loc[:, 'enddate'], y=df_po.loc[:,'adjpoll_trump'], size=10, color='red', alpha=0.03)
tab3 = Panel(child=p3, title="Both")

checkbox = CheckboxButtonGroup(labels=["Clinton", "Trump", "Johnson", "McMullin"], active=[0, 1])
show(widgetbox(checkbox))
```
